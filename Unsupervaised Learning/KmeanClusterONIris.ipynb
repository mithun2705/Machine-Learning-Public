```python
# Importing the Libraries:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# importing datasets from sklearn
from sklearn import datasets
```

```python
l = datasets.load_iris()
```

```python
df = pd.DataFrame(l['data'],columns=l.feature_names)
```

```python
df=df.drop(['sepal length (cm)','sepal width (cm)'],axis=1)
```

```python
plt.scatter(x=df['petal length (cm)'],y=df['petal width (cm)'])
```

```python
# using elbow methord to find the optimal clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,random_state=42)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('No.Of clusters')
plt.ylabel('Sum of Square Errors')
plt.show()
    
```

```python
# training the Kmean Cluster model with 3 clusters
```

```python
kmeans = KMeans(n_clusters=3,init="k-means++",random_state=42)
y_kmeans=kmeans.fit_predict(df)
```

```python
kmeans.cluster_centers_
```

```python
# ploting the Clusters and centroids

plt.scatter(x=df.iloc[y_kmeans==0,0],y=df.iloc[y_kmeans==0,1],s=100,c='red',label ='Cluster1')
plt.scatter(x=df.iloc[y_kmeans==1,0],y=df.iloc[y_kmeans==1,1],s=100,c='green',label ='Cluster2')
plt.scatter(x=df.iloc[y_kmeans==2,0],y=df.iloc[y_kmeans==2,1],s=100,c='blue',label ='Cluster3')
plt.scatter(x=kmeans.cluster_centers_[:,0],y=kmeans.cluster_centers_[:,1],s=300,c='yellow',label='Centriod')
plt.legend()
plt.show()
```
