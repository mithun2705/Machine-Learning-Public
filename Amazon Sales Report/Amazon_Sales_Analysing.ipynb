```python
# Importing Libraries 

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Imorting Datasets

df = pd.read_csv('Amazon Sales data.csv')
df.head()
```

```python
df.info()
```

```python
# Converting Dtype of Order Date,Ship Date from Object to DateTime format

df['Order Date']= pd.to_datetime(df['Order Date'])
df['Ship Date']= pd.to_datetime(df['Ship Date'])
```

```python
df.info()
```

```python
# Checking if there is any null value in the Dataset

df.isnull().sum()
```

```python
# Checking if there is any duplicate in the Datasets

df.duplicated().sum()
```

```python
df.describe()
```

```python
# Extracting Year and Month from the Order Date
df['OrderYear']=df['Order Date'].dt.year
df['OrderMonth']=df['Order Date'].dt.month
```

```python
df.head()
```

```python
monthly_sales = df.groupby(['OrderYear', 'OrderMonth'])['Total Revenue'].sum().reset_index()
monthly_sales
```

```python
yearly_sales = df.groupby('OrderYear')['Total Revenue'].sum().reset_index()
yearly_sales
```

```python
yearly_monthly_sales = df.groupby(['OrderYear','OrderMonth'])['Total Revenue'].sum().reset_index()
yearly_monthly_sales
```

```python
df.columns
```

```python
key_metrics = df[['Units Sold','Unit Price','Unit Cost','Total Revenue','Total Cost','Total Profit']]
```

```python
plt.figure(figsize=(5,3))
sns.heatmap(key_metrics.corr(),annot=True,linewidths=1)
plt.show()
```

```python
# Sales Trend Month wise
plt.figure(figsize=(7,5))
sns.lineplot(monthly_sales,x='OrderMonth',y='Total Revenue')
plt.title('Sales Trend Month-wise')
plt.xlabel('Month-Wise Sales')
plt.ylabel('Total Revenue')
plt.show()
```

```python
# Yearly Sales Trends

plt.figure(figsize=(7,5))
sns.lineplot(data=yearly_sales,x='OrderYear',y='Total Revenue')
plt.title('Yearly-Wise Sales Trends')
plt.xlabel('Yearly Sales')
plt.ylabel('Total Revenue')
plt.show()
```

```python
# Yearly month-wise sales heatmap
plt.figure(figsize=(7,5))
sns.barplot(data=yearly_monthly_sales,x='OrderMonth',y='Total Revenue')
plt.title('Sales Trend Month-wise')
plt.xlabel('Month-Wise Sales')
plt.ylabel('Total Revenue')
plt.show()
```

```python
# Visualization of unit price, cost price, and total profit
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.boxplot(y='Unit Price', data=df)
plt.title('Distribution of Unit Price')

plt.subplot(2, 2, 2)
sns.boxplot(y='Unit Cost', data=df)
plt.title('Distribution of Unit Cost')

plt.subplot(2, 2, 3)
sns.boxplot(y='Total Profit', data=df)
plt.title('Distribution of Total Profit')

plt.subplot(2, 2, 4)
sns.scatterplot(x='Unit Price', y='Total Profit', data=df)
plt.title('Total Profit vs. Unit Price')
plt.tight_layout()
plt.show()
```

```python
# Visualization of top-selling products
top_products = df.groupby('Item Type')['Units Sold'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_products.plot(kind='bar', color='skyblue')
plt.title('Top 10 Selling Products')
plt.xlabel('Product')
plt.ylabel('Total Units Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate total profit generated by each product
profit_per_product = df.groupby('Item Type')['Total Profit'].sum().sort_values(ascending=False)
```

```python
# Select top 10 profitable products
top_10_profitable_products = profit_per_product.head(10)
```

```python
# Plotting the top 10 profitable products
plt.figure(figsize=(10, 6))
top_10_profitable_products.plot(kind='bar', color='green')
plt.title('Top 10 Profitable Products')
plt.xlabel('Product')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate average unit price of each product
average_unit_price_per_product = df.groupby('Item Type')['Unit Price'].mean().sort_values(ascending=False)

# Select top 10 most expensive products
top_10_expensive_products = average_unit_price_per_product.head(10)

# Plotting the top 10 most expensive products
plt.figure(figsize=(10, 6))
top_10_expensive_products.plot(kind='bar', color='orange')
plt.title('Top 10 Most Expensive Products')
plt.xlabel('Product')
plt.ylabel('Average Unit Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate total sales and profit for each country
country_sales_profit = df.groupby('Country')[['Total Revenue', 'Total Profit']].sum().sort_values(by='Total Revenue', ascending=False)

# Plotting country-wise sales and profit
plt.figure(figsize=(17, 5))
country_sales_profit['Total Revenue'].plot(kind='bar', color='blue')
plt.title('Country-wise Total Sales')
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plotting total profit
plt.figure(figsize=(17, 5))
country_sales_profit['Total Profit'].plot(kind='bar', color='green')
plt.title('Country-wise Total Profit')
plt.xlabel('Country')
plt.ylabel('Total Profit')
plt.xticks(rotation=90)
plt.show()
```

```python
monthly_sales = df.resample('M', on='Order Date')['Total Revenue'].sum()

plt.figure(figsize=(10, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-')
plt.title('Monthly Total Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
# Sales by Product Category:
plt.figure(figsize=(10, 6))
df.groupby('Item Type')['Total Revenue'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Sales Channel Performance:
plt.figure(figsize=(8, 6))
df.groupby('Sales Channel')['Total Revenue'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales Distribution by Channel')
plt.ylabel('')
plt.show()
```

```python
# Order Priority:
plt.figure(figsize=(8, 6))
df.groupby('Order Priority')['Total Revenue'].sum().plot(kind='bar')
plt.title('Total Sales by Customer Segment')
plt.xlabel('Order Priority')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.show()
```
