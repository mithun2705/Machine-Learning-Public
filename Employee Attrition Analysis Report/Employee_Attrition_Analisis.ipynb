```python
# Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for i in df:
    if df[i].dtype == 'object':
        df[i]=le.fit_transform(df[i])
        
```

```python
df.drop(['EmployeeCount','Over18','StandardHours'],inplace=True,axis=1)
```

```python
# Correlation Analysis
correlation_matrix = df.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
df['MonthlyIncome'].value_counts()
```

```python
def incomeRange(income):
    if income<35000:
        return 'Low'
    elif income>100000:
        return 'High'
    else:
        return 'Average'
```

```python
df['IncomeRange'] = df['MonthlyIncome'].apply(incomeRange)
```

```python
df['IncomeRange']
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x='IncomeRange',hue='Attrition',data=df)
plt.title('Attrition Distribution by Monthly Income')
plt.xlabel('Monthly Income')
plt.ylabel('count')
plt.xticks(rotation=90)
plt.show()
```
