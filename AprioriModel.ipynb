```python
# importing Essential Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import datasets

df = pd.read_csv('Market_Basket_Optimisation.csv',header=None)
df
```

```python
df.shape
```

```python
transactions = []
for items in range(0,7501):
    transactions.append([str(df.values[items,j]) for j in range(0,20)])
```

```python
# Training the Apriori Model on the datasets
# ! pip install apyori
from apyori import apriori
rules = apriori(transactions=transactions,min_support = .003,min_confidence = 0.2, min_lift = 2, min_length = 2,max_length=2 )
```

```python
# Display the first results coming directly from the output of apriori
results = list(rules)
results
```

```python
# Putting the data in Pandas Dataframe
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    supports = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    return list(zip(lhs,rhs,supports,confidences,lifts))
df1 = pd.DataFrame(inspect(results),columns=['Left Hand Side','Right Hand Side','Support','Confidence','Lift'])
```
